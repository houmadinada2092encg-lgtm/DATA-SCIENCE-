```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets

# metadata
print(iris.metadata)

# variable information
print(iris.variables)
```

# Task
Perform data visualization on the Iris dataset by combining the features (X) and the target variable (y) into a single pandas DataFrame, then generating histograms for each continuous feature, box plots for each continuous feature grouped by class, and a pair plot to show relationships between all feature pairs, colored by class. Finally, present and explain the generated plots.

## Préparer les Données pour la Visualisation

### Subtask:
Combinez les caractéristiques (X) et la variable cible (y) en un seul DataFrame pour faciliter la visualisation. Cela inclura les colonnes 'sepal length', 'sepal width', 'petal length', 'petal width' et 'class'.


**Reasoning**:
Combine the features (X) and target (y) into a single DataFrame `df_iris` for easier visualization by concatenating them along the column axis.



```python
import pandas as pd

df_iris = pd.concat([X, y], axis=1)
print("Combined DataFrame head:")
print(df_iris.head())
```

## Visualisation des Distributions (Histograms)

### Subtask:
Générez des histogrammes pour chaque caractéristique continue (longueur du sépale, largeur du sépale, longueur du pétale, largeur du pétale) afin de visualiser leurs distributions. Incluez des titres et des étiquettes pour chaque graphique.


**Reasoning**:
To visualize the distributions of continuous features, I will generate histograms for each feature using `seaborn.histplot`.



```python
import matplotlib.pyplot as plt
import seaborn as sns

continuous_features = ['sepal length', 'sepal width', 'petal length', 'petal width']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(continuous_features):
    plt.subplot(2, 2, i + 1) # Create a 2x2 grid for the plots
    sns.histplot(df_iris[feature], kde=True)
    plt.title(f'Distribution of {feature.replace('_', ' ').title()}')
    plt.xlabel(feature.replace('_', ' ').title())
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Visualisation des Distributions par Classe (Box Plots)

### Subtask:
Générez des box plots pour chaque caractéristique continue, regroupées par la variable 'class', afin de visualiser la distribution de chaque caractéristique pour chaque espèce d'Iris. Incluez des titres et des étiquettes appropriés pour chaque graphique.

#### Instructions
1. Iterate through each of the continuous feature columns in `df_iris`: 'sepal length', 'sepal width', 'petal length', and 'petal width'.
2. For each feature, create a box plot using `seaborn.boxplot()`, specifying the feature as the `y`-axis and 'class' as the `x`-axis.
3. Set an appropriate title for each box plot, for example, 'Distribution of [Feature Name] by Class'.
4. Label the x-axis as 'Iris Class' and the y-axis with the feature name.
5. Ensure each plot is displayed separately or in a grid for clarity using `plt.show()` after each plot or `plt.tight_layout()` if using subplots.

**Reasoning**:
To visualize the distribution of each continuous feature across different Iris species, I will generate box plots for each feature grouped by the 'class' variable, using `seaborn.boxplot`.



```python
plt.figure(figsize=(15, 10))
for i, feature in enumerate(continuous_features):
    plt.subplot(2, 2, i + 1) # Create a 2x2 grid for the plots
    sns.boxplot(x='class', y=feature, data=df_iris)
    plt.title(f'Distribution of {feature.replace('_', ' ').title()} by Class')
    plt.xlabel('Iris Class')
    plt.ylabel(feature.replace('_', ' ').title())

plt.tight_layout()
plt.show()
```

## Visualisation des Relations entre les Caractéristiques (Pair Plot)

### Subtask:
Générez un pair plot pour visualiser les relations entre toutes les paires de caractéristiques continues, en colorant les points par la variable 'class', afin de comprendre les interactions et les séparations entre les espèces.

#### Instructions
1. Use `seaborn.pairplot()` with the `df_iris` DataFrame.
2. Specify `hue='class'` to color the points according to the Iris species.
3. Add a title to the pair plot indicating its purpose.

**Reasoning**:
To visualize the relationships between all pairs of continuous features and how these relationships differ across the Iris species, I will generate a pair plot using `seaborn.pairplot` with the `class` variable as the hue.



```python
pair_plot = sns.pairplot(df_iris, hue='class')
pair_plot.fig.suptitle('Pair Plot of Iris Features by Class', y=1.02) # Add a title to the pair plot
plt.show()
```
